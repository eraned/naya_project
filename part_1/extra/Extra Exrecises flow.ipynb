## Exercise 8

In the following equation each letter represents a different digit from 1 to 9. Can you find which letter stands for which digit?

<center> $\frac{AB}{CD}+\frac{EF}{GHI}=7$

since $\frac{EF}{GHI} \le 1$, then $\frac{AB}{CD} \ge 6$, and since $C \ge 12$, then $AB \ge 72$, and we can limit the search significantly.

### Solution

````
for n in range(721345689, 987654321):
    s_n = str(n)
    if '0' in s_n or len(set(s_n)) < 9:
        continue
    AB = int(s_n[:2])
    CD = int(s_n[2:4])
    EF = int(s_n[4:6])
    GHI = int(s_n[6:])
    if AB/CD + EF/GHI == 7:
        print(AB, CD, EF, GHI)
        break
````

## Exercise 3


*   Part 1 - Implement the 21 game, so that you can play with the computer.
The play should be interactive, and the winner should be announced.
Make sure not to go to negative numbers.
*   Part 2 - Can you implement it, so that if the computer is playing second, it will always wins?





### Solution

````
n = 21
min_take = 1
max_take = 3
````

````
from random import choice
while n:
    print(n)
    my_turn = int(input("How many coins do you want to take?\n"))
    while my_turn not in range(min_take, min(max_take, n)+1):
        my_turn = int(input(f"You must take between {min_take} and {min(max_take, n)}. Please choose again."))
    n -= my_turn
    if n==0:
        print("You win!")
        break

    print(n)
    pc_turn = choice(range(min_take, min(max_take, n)+1))
    print(f"I take {pc_turn}")
    n -= pc_turn
    if n==0:
        print("PC wins!")
        break
````
