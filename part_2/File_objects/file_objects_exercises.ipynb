# Exercise 1

* Part 1 - Make a list of words and create a file called “word length.txt”, containing a line saying “The word <word\> has <n\> letters” for each word in the list.
* Part 2 - Read the file "word_length.txt" and parse it to re-create your original list of words.

## Solution

### Part 1

```
my_words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
with open('word length.txt', 'w') as f:
    for word in my_words:
        f.write('The word {} has {} letters.\n'.format(word, len(word)))
```

### Part 2

```
your_list = []
with open('word length.txt', 'r') as f:
    for line in f:
        your_list.append(line.split()[2])
print(your_list)
```

# Exercise 2

The file customers.txt contains the information of all my customers. Read the data from the ffile into relevant data structures and answer the following questions:

1. How many customers do I have?
2. In what streets do they live?
3. How many of my customeers live on each street?
4. Which customers live on the 4th floor?

> **Note:** You can use the `Table` class we've developed in the OOP lesson

## solution

### Preliminaries

```
import sys

if 'google.colab' in sys.modules:
    from google.colab import files
    uploaded = files.upload()
```

```
class Table:
    def __init__(self, fields):
        self.fields = fields
        self.n_fields = len(fields)
        self.records = []
        self.n_records = 0
    
    def add_record(self, rec):
        self.records.append(rec)
        self.n_records += 1
        
    def remove_record(self, rec):
        self.records.remove(rec)
        self.n_records -= 1
        
    def get_column(self, field):
        ind = self.fields.index(field)
        return [rec[ind] for rec in self.records]        
                
    def select_where(self, field, value):
        ind = self.fields.index(field)
        ret = Table(self.fields)
        for rec in self.records:
            if rec[ind] == value:
                ret.add_record(rec)
        return ret
```

### Get the data

```
customers = Table(['name', 'street', 'house', 'appartment', 'floor'])
with open('customers.txt') as f:
    for line in f:
        customer = line.split(',')
        
        name = ' '.join(customer[0].split()[1:])
        street = customer[1].split()[1]
        house = int(customer[2].split()[1])
        appartment = int(customer[3].split()[1])
        floor = int(customer[4].split()[1])
        
        customers.add_record([name, street, house, appartment, floor])
```

### Question 1

```
customers.n_records
```

### Question 2

```
streets = set(customers.get_column('street'))
print(streets)
```

### Question 3

```
for street in streets:
    n_customers = customers.select_where('street', street).n_records
    print("There are {} customers on {} street.".format(n_customers, street))
```

### Question 4

```
print(customers.select_where('floor', 4).get_column('name'))
```

# Exercise 3

The file “queue.txt” contains a list of patients by their time of their arrival to a clinic. Once in a while the doctor “calls” the patient who is first in the line. Apply the class Queue (Developped in the exercise to the OOP lesson) to answer who is the patient that suffered the longest line.

## Solution

### Preliminaries

```
class Queue:
    def __init__(self):
        self.queue = []
        
    def is_empty(self):
        return len(self.queue) == 0
    
    def enqueue(self, element):
        self.queue.insert(0, element)
        
    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop()
    
    def get_len(self):
        return len(self.queue)
```

```
import sys

if 'google.colab' in sys.modules:
    from google.colab import files
    uploaded = files.upload()
```

```
waiting = []
patients = Queue()
with open("queue.txt") as f:
    for line in f:
        if "Call" not in line:
            patient = line.strip()
            patients.enqueue(patient)
            waiting.append((patients.get_len(), patient))
        else:
            patients.dequeue()
poor_guy = sorted(waiting)[-1][-1]
```
